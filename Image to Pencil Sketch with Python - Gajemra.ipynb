# Task 2 : Image to Pencil Sketch with Python

# Importing the required libraries!!

```python
import cv2
from PIL import Image
```

# Loading the Image

```python
image = cv2.imread("boy.jpg")
```

```python
image = cv2.resize(image,(400,500))
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

# Converting Image to Grey Image

```python
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

# Converting the Image to Negative Image

```python
invert_img = cv2.bitwise_not(grey_img)
display(Image.fromarray(invert_img))
```

# Smoothening the Image

```python
smoothing_img = cv2.GaussianBlur(invert_img, (5,5), sigmaX=0, sigmaY=0)
display(Image.fromarray(smoothing_img))
```

# Converting Image into Pencil Sketch

```python
sketch_img = 255 - smoothing_img
final_sketch = cv2.divide(grey_img, sketch_img, scale=230.0)
display(Image.fromarray(final_sketch))
```

# Compairing Original and Pencil Sketch Image

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
display(Image.fromarray(final_sketch))
```
